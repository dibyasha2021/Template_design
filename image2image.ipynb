```python
import torch
import requests
import torch.nn as nn
import torch.nn.functional as F
from PIL import Image
from IPython.display import display

from io import BytesIO
from tqdm.auto import tqdm
from matplotlib import pyplot as plt
from torchvision import transforms as tfms
```

```python
from diffusers.utils import make_image_grid
from diffusers import StableDiffusionXLImg2ImgPipeline
from diffusers import (
    LMSDiscreteScheduler,
    AutoPipelineForInpainting,
    StableDiffusionInpaintPipeline,
    DPMSolverMultistepScheduler,
    DDIMScheduler,
    AutoPipelineForImage2Image,
    DiffusionPipeline,
    StableDiffusionPipeline,
    StableDiffusionImg2ImgPipeline,
    StableDiffusionXLImg2ImgPipeline
    )
```

```python
# If you have a URL
# Useful function for later
def load_image(url, size=None):
    response = requests.get(url,timeout=0.2)
    img = Image.open(BytesIO(response.content)).convert('RGB')
    if size is not None:
        img = img.resize(size)
    return img
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
# Load a pipeline


pipeline = StableDiffusionXLImg2ImgPipeline    #@param

## Example for Inpainting
### StableDiffusionInpaintPipeline,AutoPipelineForInpainting

## Examples For Image2image
###  StableDiffusionImg2ImgPipeline, StableDiffusionXLImg2ImgPipeline,AutoPipelineForImage2Image

## Examples For Text2Image
### DiffusionPipeline, StableDiffusionPipeline,



model_id ="stabilityai/stable-diffusion-xl-refiner-1.0" #@param

## Example for Inpainting
### "runwayml/stable-diffusion-inpainting","kandinsky-community/kandinsky-2-2-decoder-inpaint"

## Examples For Image2image
### "stabilityai/stable-diffusion-2-1-base","runwayml/stable-diffusion-v1-5","stabilityai/stable-diffusion-xl-refiner-1.0","kandinsky-community/kandinsky-2-2-decoder",

## Examples For Text2Image
### "stable-diffusion-v1-5","stabilityai/stable-diffusion-2-1-base", "stable-diffusion-v1.4",




pipe = pipeline.from_pretrained(model_id, torch_dtype=torch.float16, variant="fp16", use_safetensors=True).to(device)

## we use enable_model_cpu_offload() and enable_xformers_memory_efficient_attention(), to save memory and increase inference speed.


#pipe.enable_model_cpu_offload()
#pipe.enable_xformers_memory_efficient_attention()
```

```python

# Run the pipeline, showing some of the available arguments


pipe_output = pipe(
    prompt = "Imagine a pristine beach with turquoise waters and a golden sunset. A large rectangular banner stands proudly amidst the scenery, with a harmonious blend of sandy yellow and oceanic blue. The banner's center is an empty space, waiting to be transformed into an eye-catching advertisement that captures the essence of seaside bliss.Leafy and bright",
    negative_prompt= "Text, words, logos,People, animals,Oversaturated, blurry, low quality,dull fashion,new image",  
    guidance_scale = 7.5,      #@param    # How strongly to follow the prompt
    num_inference_steps= 35 ,   #@param # How many steps to take  
    image = input_image, # The starting image
    strength = 0.8,#@param                  #0 for no change, 1.0 for max strength
    generator=torch.Generator(device=device).manual_seed(38) # Set up a generator for reproducibility
).images[0]


# View the resulting image:

plt.imshow(pipe_output)
plt.axis('off')
plt.show()
```
